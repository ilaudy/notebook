```python
from pylab import *
import matplotlib
%matplotlib inline  
```

```python
RDM=np.loadtxt("../RDM/KP_s128.txt")
CSF=np.loadtxt("../CSF/KP_s128.txt")
AGN=np.loadtxt("../AGN/KP_s128.txt")
```

```python
print(RDM[12])
```

```python
plot(np.log10(RDM[:,0]*1.0e10),RDM[:,2]/RDM[:,3]+1,'ko')
plot(np.log10(RDM[:,0]*1.0e10),(RDM[:,2]-RDM[:,4])/RDM[:,3]+1,'k*',mfc='none')
```

```python
plot(np.log10(RDM[:,0]*1.0e10),np.log10((RDM[:,2]-RDM[:,4])/-RDM[:,3]),'k*',mfc='none')
plot(np.log10(RDM[:,0]*1.0e10),np.log10(RDM[:,2]/-RDM[:,3]),'ko')
```

```python
plot(CSF[:,0]/RDM[:,0],(CSF[:,2]/CSF[:,3]+1)/(RDM[:,2]/RDM[:,3]+1),'rd',mec='none')
plot(AGN[:,0]/RDM[:,0],(AGN[:,2]/AGN[:,3]+1)/(RDM[:,2]/RDM[:,3]+1),'gp',mec='none')
###Comparing the halo mass, show that some groups are not well agreed, with mass different >0.5!!
###Need use 0.8 < Delta M <1.2 make some cut and get ride of these groups!
```

```python
plot(np.log10(RDM[:,0]*1.0e10),RDM[:,2]/-RDM[:,3],'ko')
plot(np.log10(CSF[:,0]*1.0e10),CSF[:,2]/-CSF[:,3],'rd')
plot(np.log10(AGN[:,0]*1.0e10),AGN[:,2]/-AGN[:,3],'g*')
xlabel(r"$\log_{10} M_{DM} \ \ [M_{\odot} h^{-1}]$",fontsize=18)
ylabel(r"$\frac{2T}{|W|}$",fontsize=18)
```

```python
plot(np.log10(CSF[:,0]*1.0e10),(CSF[:,2]/-CSF[:,3])/(RDM[:,2]/-RDM[:,3]),'rd',label="CSF")
plot(np.log10(AGN[:,0]*1.0e10),(AGN[:,2]/-AGN[:,3])/(RDM[:,2]/-RDM[:,3]),'g*',label="AGN")
plot([14.21,15.09],[1,1],'k--')
xlabel(r"$\log_{10} M_{DM} \ \ [M_{\odot} h^{-1}]$",fontsize=18)
ylabel(r"Ratio $\frac{2T}{|W|}$",fontsize=18)
legend()
```

```python
plot(np.log10(CSF[:,0]*1.0e10),(CSF[:,2])/(RDM[:,2]),'rd')
plot(np.log10(AGN[:,0]*1.0e10),(AGN[:,2])/(RDM[:,2]),'g*')
plot([14.21,15.09],[1,1],'k--')
xlabel(r"$\log_{10} M_{DM} \ \ [M_{\odot} h^{-1}]$",fontsize=18)
ylabel(r"Ratio T",fontsize=18)
```

```python
plot(np.log10(CSF[:,0]*1.0e10),(CSF[:,3])/(RDM[:,3]),'rd')
plot(np.log10(AGN[:,0]*1.0e10),(AGN[:,3])/(RDM[:,3]),'g*')
plot([14.21,15.09],[1,1],'k--')
xlabel(r"$\log_{10} M_{DM} \ \ [M_{\odot} h^{-1}]$",fontsize=18)
ylabel(r"Ratio W",fontsize=18)
```

```python
plot(np.log10(CSF[:,0]*1.0e10),(CSF[:,4]+CSF[:,5])/(RDM[:,4]),'rd')
plot(np.log10(AGN[:,0]*1.0e10),(AGN[:,4]+AGN[:,5])/(RDM[:,4]),'g*')
plot([14.21,15.09],[1,1],'k--')
ylim(0.4,1.6)
xlabel(r"$\log_{10} M_{DM} \ \ [M_{\odot} h^{-1}]$",fontsize=18)
ylabel(r"Ratio E_s",fontsize=18)
```
